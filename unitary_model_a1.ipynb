```julia
using QuadGK
using Plots
```

```julia
λ(x,y,z) = x^2+y^2+z^2-2*x*y-2*y*z-2*z*x
function f1(s)
    mρ = 0.7755;
    Γρ = 0.15;
    mπ = 0.139;
    R = 5;
    p = sqrt(λ(mπ^2,mπ^2,s)/(4*s));
    return 1./(mρ^2-s-1im*mρ*Γρ)*p/sqrt(1./R^2+p^2);
end
function f0(s)
    mσ = 0.8;
    Γσ = 0.8;
    mπ = 0.139;
    return 1./(mσ^2-s-1im*mσ*Γσ);
end
pl = plot(layout = 2)
plot!(pl[1], e->real(f1(e^2)),0.3,1.4)
plot!(pl[1], e->imag(f1(e^2)),0.3,1.4)
plot!(pl[2], e->real(f0(e^2)),0.3,1.4)
plot!(pl[2], e->imag(f0(e^2)),0.3,1.4)
```
